```python
# Comme pour les variables quantitatives, on va faire 2 fonctions pour analyser les variables qualitatives

def describe_tableau_quali(data, var):
    describe = data[var].describe()
    df = pd.DataFrame({'Caracteristiques': describe.index, 'Valeurs': describe.values})
    f = []
    for i in data[var].unique():
        f.append((f'Nombre de {i}',data.loc[data[var]==i].shape[0]))
        f.append((f'Nombre de {i} en %',(data.loc[data[var]==i].shape[0]/data.shape[0])*100))
    df2 = pd.DataFrame(f, columns=('Caracteristiques','Valeurs'))
    return pd.concat([df,df2], ignore_index=True)

def describe_graph_quali(data, var):
    fig, ax = plt.subplots()

    # Créer un countplot pour les données catégorielles
    sns.countplot(x=var, data=data, ax=ax)

    # Annoter les barres avec les valeurs en pourcentage
    total = len(data[var])
    for p in ax.patches:
        height = p.get_height()
        ax.annotate(f'{height / total * 100:.2f}%', 
                (p.get_x() + p.get_width() / 2., height), 
                ha='center', va='bottom')
    plt.xticks(rotation=90)
    return 
```

## C - Valeurs aberrantes

```python
!pip install scipy
from scipy import stats

def val_aberrantes_zscore(data, var):
    z_score=pd.DataFrame()
    z_score[f'z_{var}']= stats.zscore(data[var])
    val_aberrantes_z = z_score[z_score[f'z_{var}']>3]
    #print(f'il y a {val_aberrantes_z.shape[0]} valeurs aberrantes pour {var} selon z-score')
    #print(f'Soit {(val_aberrantes_z.shape[0]/data.shape[0])*100} % selon z-score')
    return val_aberrantes_z.shape[0],(val_aberrantes_z.shape[0]/data.shape[0])*100

def val_aberrantes_iqr(data, var):
    iqr = np.quantile(data[var], q=[0.25,0.75])
    limite_basse = iqr[0]-1.5*(iqr[1]-iqr[0])
    limite_haute = iqr[1]+1.5*(iqr[1]-iqr[0])
    val_aberrantes_iqr = data.loc[(data[var]<limite_basse)|(data[var]>limite_haute)]
    #print(f'il y a {val_aberrantes_iqr.shape[0]} valeurs aberrantes pour {var} selon IQR')
    #print(f'Soit {(val_aberrantes_iqr.shape[0]/data.shape[0])*100} % selon IQR')
    return val_aberrantes_iqr.shape[0], (val_aberrantes_iqr.shape[0]/data.shape[0])*100
```

```python
val_dico = {}
for var in application_train.select_dtypes('int','float').columns:
    zscore_aberrations = val_aberrantes_zscore(application_train, var)
    iqr_aberrations = val_aberrantes_iqr(application_train, var)
    val_dico[var] = {
        'Valeurs aberrantes Z-score': zscore_aberrations[0],
        'Valeurs aberrantes Z-score %': zscore_aberrations[1],
        'Valeurs aberrantes IQR': iqr_aberrations[0],
        'Valeurs aberrantes IQR %': iqr_aberrations[1]
    }
    
valeurs_aberrantes = pd.DataFrame.from_dict(val_dico, orient='index').reset_index()
valeurs_aberrantes.columns = ['Variables', 'Valeurs aberrantes Z-score', 'Valeurs aberrantes Z-score %',
                              'Valeurs aberrantes IQR', 'Valeurs aberrantes IQR %']

valeurs_aberrantes
```

```python
# On va maintenant faire une regression polynomiale en créant de nouvelles valeurs à la puissance 3.
# Avec les variables EXT_SOURCE_1, EXT_SOURCE_2, EXT_SOURCE_3? DAYS_BIRTH ET TARGET

# Créer un nouveau dataframe pour faire des varaibles polynomiales que sur celles-ci
poly_features = application_train[['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3', 'DAYS_BIRTH', 'TARGET']]

# imputer valeurs manquantes par leur mediane car PolynomialFeatures ne marche qu'avec des valeurs non-nulles
imputer = SimpleImputer(strategy = 'median')

poly_target = poly_features['TARGET']

poly_features = poly_features.drop(columns = ['TARGET'])

poly_features = imputer.fit_transform(poly_features)
                                  
# Créer l’objet polynomial avec le degré spécifié
poly_transformer = PolynomialFeatures(degree = 3)
# Entraîner les entités polynomiales
poly_transformer.fit(poly_features)

# Transform the features
poly_features = poly_transformer.transform(poly_features)
print('Polynomial Features shape: ', poly_features.shape)

# on donne des noms au nouvelles variables:
poly_transformer.get_feature_names_out(input_features = ['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3', 'DAYS_BIRTH'])[:15]
```

```python
# On créer la DataFrame associée
poly_features = pd.DataFrame(poly_features, 
                             columns = poly_transformer.get_feature_names(['EXT_SOURCE_1', 'EXT_SOURCE_2', 
                                                                           'EXT_SOURCE_3', 'DAYS_BIRTH']))

# On ajoute la target pour calculer les corrélations des nouvelles variables créer avec la target
poly_features['TARGET'] = poly_target
poly_correlations = poly_features.corr()['TARGET'].sort_values(ascending=False)

# Afficher les corrélations les plus fortes
print(poly_correlations.head(10))
print(poly_correlations.tail(5))
```

```python
# Fusionner polynomial features avec le training dataframe
poly_features['SK_ID_CURR'] = application_train['SK_ID_CURR']
application_train_poly = application_train.merge(poly_features, on = 'SK_ID_CURR', how = 'left')

application_train_ori = pd.read_csv('data/application_train.csv', sep=',')
print(f'Nombre de variables originales : {application_train_ori.shape[1]}, '
      f'Nombre de variables totales : {application_train_poly.shape[1]}')
print(f'On a créer : {application_train_poly.shape[1]-application_train_ori.shape[1]} nouvelles variables')
```

```python
le = LabelEncoder()
le_count = 0

for col in application_train_poly:
    if application_train_poly[col].dtype == 'object':
        if len(list(application_train_poly[col].unique())) <= 2:
            le.fit(application_train_poly[col])
            # Transformation de training  data
            application_train_poly[col] = le.transform(application_train_poly[col])
            le_count += 1
            
print(f'On a encoder avec label encoding {le_count} variables')
```
